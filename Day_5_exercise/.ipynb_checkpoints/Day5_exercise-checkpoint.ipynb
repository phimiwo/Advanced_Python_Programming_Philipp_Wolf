## 1. Scipy

### Linear algebra

```python
import scipy.linalg
import numpy as np

A = np.arange(1,10).reshape(3,3)
b = np.arange(1,4)
print("A = ", A)
print("b = ", b)
```

```python
x = scipy.linalg.solve(A,b)
print(x)
```

```python
b_test = np.matmul(A,x) 
print(b_test)
```

```python
B = np.random.rand(9).reshape(3,3)
print(B)
```

```python
x_2 = scipy.linalg.solve(A,B)
print(x_2)
```

```python
B_test = np.matmul(A,x_2) 
print(B_test)
#results are quite far of, probably as results of rounding errors
```

```python
## Eigenvalue problem

eigen_value_of_A, eigen_vector_of_A = scipy.linalg.eig(A)
print('Eigenvalues: ',eigen_value_of_A)
print('Eigenvectors: ', eigen_vector_of_A)
```

```python
# Calculate the inverse, determinant of A
inv_det_of_A = scipy.linalg.det(scipy.linalg.inv(A))
print(inv_det_of_A)
```

```python
# Calculate the norm of A with different orders
norm_none = scipy.linalg.norm(A)
print('Norm none: ', norm_none)

norm_1 = scipy.linalg.norm(A,ord=1)
print('Norm 1: ', norm_1)

norm_2 = scipy.linalg.norm(A,ord=2)
print('Norm 2: ', norm_2)

norm_m1 = scipy.linalg.norm(A,ord=-1)
print('Norm -1: ', norm_m1)

norm_m2 = scipy.linalg.norm(A,ord=-2)
print('Norm -2: ', norm_m2)

norm_inf = scipy.linalg.norm(A,ord=np.inf)
print('Norm inf: ', norm_inf)
```

### Statistics

```python
import scipy.stats
import matplotlib.pyplot as plt
```

```python
###POISSON
mu = 50.
```

```python
%matplotlib notebook
```

```python
fig, ax = plt.subplots(1, 1)
x = np.arange(scipy.stats.poisson.ppf(0.01, mu), scipy.stats.poisson.ppf(0.99, mu))
ax.plot(x, scipy.stats.poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
x = np.arange(scipy.stats.poisson.ppf(0.01, mu), scipy.stats.poisson.ppf(0.99, mu))
ax.plot(x, scipy.stats.poisson.cdf(x, mu), 'bo', ms=8, label='poisson cdf')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
r = scipy.stats.poisson.rvs(mu, size=1000)
ax.hist(r)
plt.show
```

```python
###NORMAL
fig, ax = plt.subplots(1, 1)
x = np.arange(scipy.stats.norm.ppf(0.01, mu), scipy.stats.norm.ppf(0.99, mu))
ax.plot(x, scipy.stats.norm.pdf(x, mu), 'bo', ms=8, label='Normal pdf')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
x = np.arange(scipy.stats.norm.ppf(0.01, mu), scipy.stats.norm.ppf(0.99, mu))
ax.plot(x, scipy.stats.norm.cdf(x, mu), 'bo', ms=8, label='Normal cdf')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
r = scipy.stats.norm.rvs(mu, size=1000)
ax.hist(r,label='Normal distr')
ax.legend()
plt.show
```

```python
rvs1 = scipy.stats.norm.rvs(loc=5,scale=10,size=500)
rvs2 = scipy.stats.norm.rvs(loc=5,scale=10,size=500)
scipy.stats.ttest_ind(rvs1,rvs2)
```
